```python
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("co2temp.csv")

temperature = np.array(data["Degrees C"]).reshape(-1,1)
carbon = np.array(data["PPM"]).reshape(-1,1)

model = LinearRegression().fit(temperature,carbon)
print(model.coef_)
print(model.intercept_)

fig = plt.figure()
fig.set_facecolor("white")
ax = fig.add_subplot(111)
ax.grid(True)
ax.scatter(temperature, carbon, label="data scatter")
# ax.plot(temperature, 88.209*temperature + 326.367, color="green")
x = [1.06981,1.220139,1.37046,1.74627,2.122085]
y = [438.938,463.8487,488.759,551.0353,613.3114]
ax.plot([-0.25,2.2],[304.3,540.4],color="green",label="regression line")
ax.scatter(x,y,label="Predictions by ARIMA")
x1 = [1.036467338,1.21294639,1.391232038,1.844973876,2.310419897]
y1 = [427.5933066,447.6470773,468.6413187,525.5361057,589.3382011]
ax.scatter(x1,y1,label="Predictions by grey forecast")
ax.text(0.85,445,"2030")
ax.text(0.97,465,"2040")
ax.text(1.13,490,"2050")
ax.text(1.5,551,"2075")
ax.text(1.9,615,"2100")
plt.xlim([-0.25,2.35])
plt.ylim([300,650])
plt.xlabel("Temperature change / degrees C")
plt.ylabel("PPM")
plt.legend(loc="best")
plt.savefig("ppmdegrees.png",dpi=500)
```
